```python
nhl_df2018['team_lastword']=nhl_df2018.team.str.strip('*').str.split(' ').str[-1]
```

```python
nhl_df2018.head()
```

```python
nhl_df2018.drop(0,inplace=True)
```

```python
nhl_df2018.team = nhl_df2018.team.str.strip('*')
```

```python
nhl_df2018.drop([9,18,26], inplace=True)
```

```python
nhl_df2018 = nhl_df2018[['team','W', 'L', 'team_lastword']]
nhl_df2018.head()
```

```python
cities.head()
```

```python
cities.shape
```

```python
cities.NHL = cities.NHL.replace("\[\w.*\]|—",'',regex=True).apply(lambda x: re.findall('[A-Z][^A-Z]*', x))
```

```python
cities.head()
```

```python
cities = cities[['NHL','Metropolitan area','Population (2016 est.)[8]']]
cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
cities.head()
```

```python
cities.NHL
```

```python
cities.at[34,'NHL'] = ['Jackets']
```

```python
cities.at[43,'NHL'] = ['Knights']
```

```python
cities.at[12,'NHL'] = ['Wings']
```

```python
cities.at[13,'NHL'] = ['Leafs']
```

```python
cities.NHL = cities.NHL.str[0]
```

```python
merged_NHL = pd.merge(nhl_df2018, cities, how='left', left_on='team_lastword', right_on='NHL')
merged_NHL
```

```python
merged_NHL.at[12, ['Metropolitan area','Population']] = ['New York City', 20153634]
merged_NHL.at[14, ['Metropolitan area','Population']] = ['New York City', 20153634]
merged_NHL.at[24, ['Metropolitan area','Population']] = ['Los Angeles', 13310447]
```

```python
merged_NHL
```

```python
merged_NHL.info()
```

```python
merged_NHL.Population = merged_NHL.Population.astype('int')
merged_NHL.W = merged_NHL.W.astype('int')
merged_NHL.L = merged_NHL.L.astype('int')
```

```python
metro_group = merged_NHL.groupby('Metropolitan area').agg({'W':np.sum, 'L':np.sum, 'Population':np.mean})
metro_group
```

```python
metro_group['W/L Rate'] = metro_group.W / (metro_group.W + metro_group.L)
metro_group
```

```python
metro_group['W/L Rate'].corr(metro_group.Population)
```
