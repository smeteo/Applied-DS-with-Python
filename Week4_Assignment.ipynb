```python
cities.NHL = cities.NHL.replace("\[\w.*\]|—",'',regex=True).apply(lambda x: re.findall('[A-Z][^A-Z]*', x))
```

```python
cities.at[34,'NHL'] = ['Jackets']
cities.at[43,'NHL'] = ['Knights']
cities.at[12,'NHL'] = ['Wings']
cities.at[13,'NHL'] = ['Leafs']
```

```python
merged_NHL
```

```python
merged_NHL.info()
```

```python
merged_NHL.Population = merged_NHL.Population.astype('int')
merged_NHL.W = merged_NHL.W.astype('int')
merged_NHL.L = merged_NHL.L.astype('int')
```

```python
metro_group = merged_NHL.groupby('Metropolitan area').agg({'W':np.sum, 'L':np.sum, 'Population':np.mean})
metro_group
```

```python
metro_group['W/L Rate'] = metro_group.W / (metro_group.W + metro_group.L)
metro_group
```

```python
metro_group['W/L Rate'].corr(metro_group.Population)
```
