```python
cities = cities[['NHL','Metropolitan area','Population (2016 est.)[8]']]
cities.rename(columns={'Population (2016 est.)[8]':'Population'},inplace=True)
cities.head()
```

```python
cities.at[34,'NHL'] = ['Jackets']
cities.at[43,'NHL'] = ['Knights']
cities.at[12,'NHL'] = ['Wings']
cities.at[13,'NHL'] = ['Leafs']
```

```python
cities.NHL = cities.NHL.str[0]
```

```python
merged_NHL = pd.merge(nhl_df2018, cities, how='left', left_on='team_lastword', right_on='NHL')
merged_NHL
```

```python
metro_group = merged_NHL.groupby('Metropolitan area').agg({'W':np.sum, 'L':np.sum, 'Population':np.mean})
metro_group
```

```python
metro_group['W/L Rate'] = metro_group.W / (metro_group.W + metro_group.L)
metro_group
```

# Question 4

Calculate the win/loss ratio's correlation with the population of the city it is in for the NFL using 2018 data.

```python
nfl_df=pd.read_csv("nfl.csv")
cities=pd.read_html("wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python
nfl_df2018 = nfl_df[nfl_df.year == 2018]
nfl_df2018.team = nfl_df2018.team.replace('\*|\+','',regex=True)
nfl_df2018['team_lastword'] = nfl_df2018.team.str.split(' ').str[-1]
nfl_df2018 = nfl_df2018[['team','W-L%', 'team_lastword']]
nfl_df2018.team_lastword = nfl_df2018.team_lastword.astype(str).str.strip()
nfl_df2018.drop([0,5,10,15,20,25,30,35],inplace=True)
nfl_df2018
```

```python
cities
```

```python
cities = cities[['NFL','Metropolitan area','Population (2016 est.)[8]']]
cities.rename(columns={'Population (2016 est.)[8]':'Population', 'Metropolitan area':'Metro'},inplace=True)
cities.NFL = cities.NFL.replace("\[\w.*\]|—",'',regex=True).apply(lambda x: re.findall('[A-Z][^A-Z]*', x))
cities.NFL = cities.NFL.str[0]
cities.NFL = cities.NFL.str.strip()
cities.NFL =cities.NFL.astype(str).str.strip()
cities
```

```python
merged_NFL = pd.merge(nfl_df2018, cities, how='left', left_on='team_lastword', right_on='NFL')
merged_NFL
```

```python
merged_NFL.at[3, ['Metro','Population']] = ['New York City',20153634]
merged_NFL.at[13, ['Metro','Population']] = ['Los Angeles',13310447]
merged_NFL.at[30, ['Metro','Population']] = ['San Francisco Bay Area', 6657982]
merged_NFL['W-L%'] = merged_NFL['W-L%'].astype(float)
merged_NFL.Population = merged_NFL.Population.astype(int)
merged_NFL = merged_NFL.groupby('Metro').mean()
merged_NFL
```

```python
merged_NFL['W-L%'].corr(merged_NFL.Population)
```
